```python
gdf.to_file("data/Zensus2022_100m_poly_GER_wPLZ_wRS_ew.fgb", driver="FlatGeobuf")
```

```python
def fgb_to_pmtiles_layer( polys_fgb, output_pmtiles, layer_name):
    import subprocess
    from pathlib import Path

    subprocess.run([
        "tippecanoe", "-o", str(Path(output_pmtiles).resolve()),
        f"--layer={layer_name}-polys",
        "--minimum-zoom=9", "--maximum-zoom=10",
        "--force",
        "--no-feature-limit", "--no-tile-size-limit", "--force-feature-limit",
        str(Path(polys_fgb).resolve())
    ], check=True)


    print("✅ PMTiles created")
```

```python
fgb_to_pmtiles_layer(
    polys_fgb="data/Zensus2022_100m_poly_GER_wPLZ_wRS_ew.fgb",
    output_pmtiles="data/Zensus2022_100m_poly_GER_wPLZ_wRS_ew_10.pmtiles",
    layer_name="rasters"
    )
```
